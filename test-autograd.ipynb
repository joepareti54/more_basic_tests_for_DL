```python
import torch
torch.manual_seed(0)
x = torch.rand(3)
#b = torch.rand([10, 1, 1, 1])
print(x)
```

```python
x.requires_grad_(True)
#print(x, x.dtype())
print(x)
```

```python
y = x + 2
print(y)
```

```python
z = y*y
print(z)
```

```python
Z= z.mean()
#Z = z
print(Z)
```

```python
Z.backward()
print(x.grad)
```
