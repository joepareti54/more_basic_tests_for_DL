```python
import numpy as np
np.random.seed(1)
x = np.random.rand(4)
#
print(x)
y = 2*x
print(y)
```

```python
def loss(y, y_hat):
    l= (y- y_hat)**2
    return(l.mean())
```

```python
def compute_grad(w, x):
    grad=2*(w*x-2*x)*x
    grad = 2* np.dot((w*x-2*x), x)
    return(grad.mean())
```

```python
def forward_pass(x, w):
    y_hat= w*x
    return(y_hat)
```

```python
alpha = 0.01
epochs = 1000
w = 0.0
```

```python
for i in range(epochs):
    y_hat = forward_pass(x,w)
    l = loss(y, y_hat)
    dw = -compute_grad(w, x)*alpha
    w = w + dw
    if i % 20 == 0:
        print(l, w)
```

```python
y_pred = w * x
print(y_pred)
print(y)
```
